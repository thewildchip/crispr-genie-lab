# 0 Data Loading for Efficiency Model

This notebook loads the CRISPR gRNA dataset and performs basic inspection.


```python
# IMPORTS
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# LOADING IN

FILE_PATH = "https://raw.githubusercontent.com/VKonstantakos/CRISPR-Deep-Learning/refs/heads/main/Data/2.%20Training%20datasets/1.%20DeepCRISPR-CNN-CRNNCrispr.csv"

df = pd.read_csv(FILE_PATH)
```

```python
df["direction"] = df["direction"].map({"+": 1, "-": 0})
df["direction"]
```

```python
def onehote(sequence):
    mapping = {"A": 0, "C": 1, "G": 2, "T": 3}
    seq2 = [mapping[i] for i in sequence]
    return np.eye(4)[seq2]

dna='ATTTACGGATTGCTGA'
#calling onehote function
oneHotEncodedDna= onehote(dna)
print(oneHotEncodedDna)
```

```python
for dna in df["23-nt sequence"]:
  oneHotEncodedDna = onehote(dna)
  print(oneHotEncodedDna)
```
